# Data Preprocessing

```python
#Split DataFrame in train and test
```

```python
x = df.drop(["target"],axis=1)
x.head()
```

```python
y = df['target']
y.head()
```

```python
# split dataset into train and test

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2,random_state=5)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
#Converting different units and magnitude data in one unit.
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
xtrain_sc = sc.fit_transform(xtrain)
xtest_sc = sc.transform(xtest)
```

# Breast Cancer Detection Machine Learning Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# Support vector Classifier

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(xtrain,ytrain)
ypred_svc = svc.predict(xtest)
accuracy_score(ytest, ypred_svc)
```

```python
# Train with Standard scaled Data
svc_classifier = SVC()
svc_classifier.fit(xtrain_sc, ytrain)
ypred_svc_sc = svc_classifier.predict(xtest_sc)
accuracy_score(ytest, ypred_svc_sc)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
ypred_lr = lr.predict(xtest)
accuracy_score(ytest,ypred_lr)
```

```python
# Train with Standard scaled Data
lr_classifier = LogisticRegression(random_state = 51, penalty = 'l2')
lr_classifier.fit(xtrain_sc, ytrain)
ypred_lr_sc = lr_classifier.predict(xtest_sc)
accuracy_score(ytest, ypred_lr_sc)
```

# K â€“ Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric="minkowski",p=2)
knn.fit(xtrain,ytrain)
ypred_knn = knn.predict(xtest)
accuracy_score(ytest,ypred_knn)
```

```python
#train with standard scaled data
knn_classifier= KNeighborsClassifier(n_neighbors=5, metric="minkowski",p=2)
knn_classifier.fit(xtrain_sc,ytrain)
ypred_knn_sc = knn_classifier.predict(xtest_sc)
accuracy_score(ytest,ypred_knn_sc)
```

# Naive Bayes Classifer

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(xtrain,ytrain)
ypred_nb = nb.predict(xtest)
accuracy_score(ytest,ypred_nb)
```

```python
# Train with Standard scaled Data
nb_classifier = GaussianNB()
nb_classifier.fit(xtrain_sc, ytrain)
ypred_nb_sc = nb_classifier.predict(xtest_sc)
accuracy_score(ytest, ypred_nb_sc)
```

# Decision Tree Classifier

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt_classifier.fit(xtrain, ytrain)
ypred_dt = dt_classifier.predict(xtest)
accuracy_score(ytest, ypred_dt)
```

```python
# Train with Standard scaled Data
dt_classifier2 = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt_classifier2.fit(xtrain_sc, ytrain)
ypred_dt_sc = dt_classifier.predict(xtest_sc)
accuracy_score(ytest, ypred_dt_sc)
```

# Random Forest Classifier

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier.fit(xtrain, ytrain)
ypred_rf = rf_classifier.predict(xtest)
accuracy_score(ytest, ypred_rf)
```

```python
# Train with Standard scaled Data
rf_classifier2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier2.fit(xtrain_sc, ytrain)
ypred_rf_sc = rf_classifier.predict(xtest_sc)
accuracy_score(ytest, ypred_rf_sc)
```

# Adaboost Classifier

```python

# Adaboost Classifier
from sklearn.ensemble import AdaBoostClassifier
adb_classifier = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'entropy', random_state = 200),
                                    n_estimators=2000,
                                    learning_rate=0.1,
                                    algorithm='SAMME.R',
                                    random_state=1,)
adb_classifier.fit(xtrain, ytrain)
ypred_adb = adb_classifier.predict(xtest)
accuracy_score(ytest, ypred_adb)
```

```python
# Train with Standard scaled Data
adb_classifier2 = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'entropy', random_state = 200),
                                    n_estimators=2000,
                                    learning_rate=0.1,
                                    algorithm='SAMME.R',
                                    random_state=1,)
adb_classifier2.fit(xtrain_sc, ytrain)
ypred_adb_sc = adb_classifier2.predict(xtest_sc)
accuracy_score(ytest, ypred_adb_sc)
```

# XGBoost Classifier

```python
# XGBoost Classifier
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier()
xgb_classifier.fit(xtrain, ytrain)
ypred_xgb = xgb_classifier.predict(xtest)
accuracy_score(ytest, ypred_xgb)
```

```python
# Train with Standard scaled Data
xgb_classifier2 = XGBClassifier()
xgb_classifier2.fit(xtrain_sc, ytrain)
ypred_xgb_sc = xgb_classifier2.predict(xtest_sc)
accuracy_score(ytest, ypred_xgb_sc)
```
